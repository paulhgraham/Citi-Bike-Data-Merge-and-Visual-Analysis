```python
import pandas as pd
import os
import glob
from datetime import date
```

```python
#import one csv file to test
df=pd.read_csv('../project1_data/201608-citibike-tripdata.csv', nrows = 10)
df
```

```python
#trim the excess files
path = '../project1_data/'
os.chdir(path)
files = [i for i in glob.glob('*.csv')]
files
```

```python
all_df = pd.DataFrame()
for file in files:
    file = '../project1_data/' + file
    df=pd.read_csv(file, nrows=1)
#     print(df.columns)
    all_df = all_df.append(df)
```

```python
columns = df.columns
```

```python
#clean up data table

#convert gender
# 0 = unknown, 1=male, 2=female
gender = {1:'f', 2:'m', 0:'unknown'}
all_df['gender'] = all_df['gender'].map(gender)

#generate age
def calc_age(birth_year):
    today = date.today()
    return today.year - birth_year
all_df['age'] = all_df['birth year'].apply(calc_age)
```

```python
all_df 
```

```python
#age grouping
age_bins = [0, 20, 30, 40, 50, 60, 100]
age_groups = ['0-20', '21-30', '31-40', '41-50', '51-60', '61+']
all_df['age groups'] = pd.cut(all_df['age'], age_bins, labels=age_groups)
```

```python
#new limited dataframe
bike_data = all_df[['age', 'age groups', 'gender', 'usertype','starttime', 'start station latitude', 'start station longitude', 
                   'stoptime', 'end station latitude', 'end station longitude', 'tripduration']].copy()
bike_data
```

```python
#gender count
gender = bike_data['gender'].value_counts().reset_index()
gender
```

```python
#age
age = bike_data['age'].describe()
print(age)

average_age = bike_data['age'].mean()
print('')
print('the average user age is: '+ str(average_age))

#age bin counts
bike_data['age groups'].value_counts().reset_index()
```
